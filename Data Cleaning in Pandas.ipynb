```python
# List the unique countries.
df_1990_2021['country'].unique()
```

```python
# Identify the countries that contained strings from the 'east_asia' list and place them in a list.
countries_list = []
east_asia = ['Brunei', 'Cambodia', 'China', 'Fiji', 'Indonesia', 'Kiribati', 'Laos', 'Malaysia', 'Marshall', 'Micronesia', 'Mongolia', 
             'Myanmar', 'Palau', 'Papua', 'Philippines', 'Samoa', 'Singapore', 'Solomon', 'Thailand', 'Timor', 'Tonga', 'Tuvalu', 
             'Vanuatu', 'Vietnam']

for x in east_asia:
   countries_list.append(df_1990_2021[df_1990_2021['country'].str.contains(x)]['country'].unique())
countries_list
```

```python
# Separate all the elements from the nested lists and put it into another list.
countries_list_official = []

for x in countries_list:
    for y in x:
        countries_list_official.append(y)
countries_list_official
```

```python
# Determine the number of elements inside the list.
len(countries_list_official)
```

```python
# Remove the two countries from the list which should not be included in the list of East Asian countries.
countries_list_remove = ['Asia (excl. China and India)', 'American Samoa']
countries_list_official = [x for x in countries_list_official if x not in countries_list_remove]
countries_list_official
```

```python
# Modify the dataframe to only consist of data from the countries specified on the list. 
df_1990_2021 = df_1990_2021.loc[df_1990_2021['country'].apply(lambda x: x in countries_list_official)]
df_1990_2021.head(50)
```

```python
# # Check the sum of null values per year and plot. 
df_1990_2021.groupby('year')['count_nulls'].sum().plot(kind='barh')
```
